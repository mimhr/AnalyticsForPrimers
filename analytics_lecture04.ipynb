## 不定積分(p.78,79)

### 教科書での定義

+ 原始関数F(x)の定義

$$
F'(x)=\frac{d}{dx}F(x)=f(x)
$$

原始関数$F(x)$は、$F(x)+C$と任意の定数$C\in\mathbb{R}or\mathbb{C}$も含まれる。線形代数の言葉でいうと原始関数$F(x)$は、$F(x)/C,C\in\mathbb{R}or\mathbb{C}$なる商空間をなす、ともいえる?

というのも、$G(x)=F(x)+C,C\in\mathbb{R}or\mathbb{C}$として
$$
G'(x)=\frac{d}{dx}G(x)=\frac{d}{dx}\left(F(x)+C\right)=\frac{d}{dx}F(x)\overset{\mathrm{上の定義と同じように}}{=}f(x)
$$

不定積分は$\int dx$を左側につけて、
$$
\int dx F'(x)=\int dx \frac{d}{dx}\left(F(x)+C\right)=F(x)+C=\int dx f(x)=\int f(x)dx
$$

### こういうのもあり
というより、厳密には定積分が定義された上で積分範囲を変数とすることで、不定積分を定義するのが正式(だと思っています)。

$$
F(x)=\int_{c}^{x}f(t)dt
$$

#### 微積分学の基本定理(公式)
$f(x)$が閉区間$I=[a,b]$で微分可能で$f'(x)$が$I$で連続ならば
$$
\int_{a}^{b}f'(x)dx=f(b)-f(a)
$$

不定積分は微分の"**逆演算**"として進めます。

### 基本的性質(p.79)
$$
\begin{eqnarray}
\frac{d}{dx}\int f(x)dx&=&f(x) \\
\int F'(x)dx&=&F(x)+C \\
\int k f(x)dx&=&k\int f(x)dx,\quad kは定数 \\
\int(f+g)dx&=&\int fdx+\int gdx
\end{eqnarray}
$$

### 部分積分法

$$
\begin{eqnarray}
\int e^{f(x)}f'(x)dx&=&e^{f(x)} \\
\int\frac{f'(x)}{f(x)}dx&=&\log\left|f(x)\right| \\
\int f(x)e^{x}dx&=&e^{x}(f(x)-f'(x)+f''(x)-f'''(x)...) \\
\int f(x)e^{-x}dx&=&e^{-x}(-f(x)-f'(x)-f''(x)-f'''(x)...) \\
\int f(x)\cos x dx&=&f(x)\sin x+f'(x)\cos x+f''(x)(-\sin x)+f'''(x)(-\cos x)+... \\
\int f(x)\sin x dx&=&f(x)(-\cos x)+f'(x)\sin x+f''(x)\cos x-f'''(x)\cos x+... \\
\int f(x)\log x dx&=&F(x)\log x-\int\frac{1}{x}F(x)dx,\quad F(x)=\int f(x)dx \\
\end{eqnarray}
$$


#### 例:$\displaystyle \int x^3e^{-x}dx$
上式で$f(x)=x^3$として、$f'=3x^2$、$f''=6x$、$f'''=6$、$f^{(4)}=0$となって、
$$
\begin{eqnarray}
\int x^3e^{-x}dx&=&e^{-x}(-f(x)-f'(x)-f''(x)-f'''(x)...) \\
&=&e^{-x}(-x^3-3x^2-6x-6+0+0...)
\end{eqnarray}
$$

#### 例:$\displaystyle \int x^3\cos x dx$
上式で$f(x)=x^3$として、$f'=3x^2$、$f''=6x$、$f'''=6$、$f^{(4)}=0$となって、
$$
\begin{eqnarray}
\int x^3\cos x dx&=&f(x)\sin x+f'(x)\cos x+f''(x)(-\sin x)+f'''(x)(-\cos x) \\
&=&x^3\sin x+3x^2\cos x+6x(-\sin x)+6(-\cos x)+0+0...
\end{eqnarray}
$$
