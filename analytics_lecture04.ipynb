## 不定積分(p.78,79)

### 教科書での定義

+ 原始関数F(x)の定義

$$
F'(x)=\frac{d}{dx}F(x)=f(x)
$$

原始関数$F(x)$は、$F(x)+C$と任意の定数$C\in\mathbb{R}or\mathbb{C}$も含まれる。線形代数の言葉でいうと原始関数$F(x)$は、$F(x)/C,C\in\mathbb{R}or\mathbb{C}$なる商空間をなす、ともいえる?

というのも、$G(x)=F(x)+C,C\in\mathbb{R}or\mathbb{C}$として
$$
G'(x)=\frac{d}{dx}G(x)=\frac{d}{dx}\left(F(x)+C\right)=\frac{d}{dx}F(x)\overset{\mathrm{上の定義と同じように}}{=}f(x)
$$

不定積分は$\int dx$を左側につけて、
$$
\begin{eqnarray}
\int dx F'(x)&=&\int dx \frac{d}{dx}F(x)=\int 1\cdot dF=F(x)+C \\
\int dx f(x)&=&\int f(x)dx
\end{eqnarray}
$$
より、
$$
\int f(x)dx=F(x)+C
$$

### こういうのもあり
というより、厳密には定積分が定義された上で積分範囲を変数とすることで、不定積分を定義するのが正式(だと思っています)。つまり、
$$
F(x)=\int_{c}^{x}f(t)dt
$$
とすると、
$$
\int_{c}^{x}f(t)dt=F(x)-F(c)
$$

#### 微積分学の基本定理(公式)
$f(x)$が閉区間$I=[a,b]$で微分可能で$f'(x)$が$I$で連続ならば
$$
\int_{a}^{b}f'(x)dx=f(b)-f(a)
$$

不定積分は微分の"**逆演算**"として進めます。

### 基本的性質(p.79)
$$
\begin{eqnarray}
\frac{d}{dx}\int f(x)dx&=&f(x) \\
\int F'(x)dx&=&F(x)+C \\
\int k f(x)dx&=&k\int f(x)dx,\quad kは定数 \\
\int(f+g)dx&=&\int fdx+\int gdx
\end{eqnarray}
$$

## 問題追加

$$
\int\frac{dx}{a+b\cos x},\quad (a>b>0)
$$

問題2-2.3の(1)〜(3)を使います。ここでも実際に計算してみます。

$t=\tan\frac{x}{2}$とおくと
$$
dt=\frac{1}{2}\frac{1}{\cos^2\frac{x}{2}}dx=\frac{1}{2}(1+\tan^2\frac{x}{2})dx=\frac{1}{2}(1+t^2)dx
$$
また数学公式(p.207 三角関数(4)の倍角の公式)を使うことで
$$
\begin{eqnarray}
\sin x&=&2\sin\frac{x}{2}\cos\frac{x}{2}=2\frac{\sin\frac{x}{2}}{\cos\frac{x}{2}}\cos^2\frac{x}{2}=2\tan\frac{x}{2}\cos^2\frac{x}{2}=\frac{2\tan\frac{x}{2}}{1+\tan^2\frac{x}{2}}=\frac{2t}{1+t^2} \\
\cos x&=&\cos^2\frac{x}{2}-\sin^2\frac{x}{2}=\cos^2\frac{x}{2}-(1-\cos^2\frac{x}{2})=2\cos^2\frac{x}{2}-1=\frac{2}{1+\tan^2\frac{x}{2}}-1 \\
&=&\frac{2}{1+t^2}-1=\frac{2-(1+t^2)}{1+t^2}=\frac{1-t^2}{1+t^2}
\end{eqnarray}
$$

これらを使うと、
$$
\begin{eqnarray}
I&=&\int\frac{dx}{a+b\cos x}=\int\frac{\frac{2dt}{1+t^2}}{a+b\frac{1-t^2}{1+t^2}}=\int\frac{2dt}{(a+b)+(a-b)t^2} \\
ここでt&=&\sqrt{\frac{a+b}{a-b}}uとおくと、dt=\sqrt{\frac{a+b}{a-b}}duとなり、 \\
&=&\frac{2}{a+b}\sqrt{\frac{a+b}{a-b}}\int\frac{du}{1+u^2} \\
&=&\frac{2}{\sqrt{a^2-b^2}}\tan^{-1}u \\
&=&\frac{2}{\sqrt{a^2-b^2}}\tan^{-1}\left(\sqrt{\frac{a-b}{a+b}}t\right) \\
&=&\frac{2}{\sqrt{a^2-b^2}}\tan^{-1}\left(\sqrt{\frac{a-b}{a+b}}\tan\frac{x}{2}\right)
\end{eqnarray}
$$

## 公式に付け足してください

$$
\int\sqrt{x^2+b}dx=\frac{1}{2}\left(x\sqrt{x^2+b}+b\log\left|x+\sqrt{x^2+b}\right|\right)
$$
これを出すのは、かなり面倒。
$$
\begin{eqnarray}
I&=&\int\sqrt{x^2+b}dx=\int(x)'\sqrt{x^2+b}dx,\quad部分積分の積分側\int x((x^2+b)^{1/2})'dx=\int x\frac{2x}{2}(x^2+b)^{-1/2}dx \\
&=&x\sqrt{x^2+b}-\int\frac{x^2}{\sqrt{x^2+b}}dx=x\sqrt{x^2+b}-\int\frac{x^2+b-b}{\sqrt{x^2+b}}dx \\
&=&x\sqrt{x^2+b}-\int\frac{x^2+b}{\sqrt{x^2+b}}dx+\int\frac{b}{\sqrt{x^2+b}}dx \\
&=&x\sqrt{x^2+b}-\int\sqrt{x^2+b}dx+\int\frac{1}{\sqrt{x^2+b}}dx \\
&=&x\sqrt{x^2+b}-I+\int\frac{1}{\sqrt{x^2+b}}dx
\end{eqnarray}
$$
より
$$
2I=x\sqrt{x^2+b}+\int\frac{1}{\sqrt{x^2+b}}dx
$$
となり、$\displaystyle \int\frac{1}{\sqrt{x^2+b}}dx=\log\left|x+\sqrt{x^2+b}\right|$を用いて得られる。

### 部分積分法

$$
\begin{eqnarray}
\int e^{f(x)}f'(x)dx&=&e^{f(x)} \\
\int\frac{f'(x)}{f(x)}dx&=&\log\left|f(x)\right| \\
\int f(x)e^{x}dx&=&e^{x}(f(x)-f'(x)+f''(x)-f'''(x)...) \\
\int f(x)e^{-x}dx&=&e^{-x}(-f(x)-f'(x)-f''(x)-f'''(x)...) \\
\int f(x)\cos x dx&=&f(x)\sin x+f'(x)\cos x+f''(x)(-\sin x)+f'''(x)(-\cos x)+... \\
\int f(x)\sin x dx&=&f(x)(-\cos x)+f'(x)\sin x+f''(x)\cos x-f'''(x)\sin x+... \\
\int f(x)\log x dx&=&F(x)\log x-\int\frac{1}{x}F(x)dx,\quad F(x)=\int f(x)dx \\
\end{eqnarray}
$$


#### 例:$\displaystyle \int x^3e^{-x}dx$
上式で$f(x)=x^3$として、$f'=3x^2$、$f''=6x$、$f'''=6$、$f^{(4)}=0$となって、
$$
\begin{eqnarray}
\int x^3e^{-x}dx&=&e^{-x}(-f(x)-f'(x)-f''(x)-f'''(x)...) \\
&=&e^{-x}(-x^3-3x^2-6x-6+0+0...)
\end{eqnarray}
$$

#### 例:$\displaystyle \int x^3\cos x dx$
上式で$f(x)=x^3$として、$f'=3x^2$、$f''=6x$、$f'''=6$、$f^{(4)}=0$となって、
$$
\begin{eqnarray}
\int x^3\cos x dx&=&f(x)\sin x+f'(x)\cos x+f''(x)(-\sin x)+f'''(x)(-\cos x) \\
&=&x^3\sin x+3x^2\cos x+6x(-\sin x)+6(-\cos x)+0+0...
\end{eqnarray}
$$
