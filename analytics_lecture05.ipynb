### 分布関数

分布関数$F(x)$が連続で、(有限個の点を除けば)$F(x)$の導関数
$$
F(x)=\frac{dF(x)}{dx}\equiv f(x)
$$

が存在するとき、Xは連続な確率変数とよばれ、分布関数と確率変数は

$$
F(x)=P(X\le x)=\int_{-\infty}^{x}f(t)dt,\quad -\infty<x<\infty
$$

を満たす。ここで$f(t)$は$t$における$X$の確率密度関数(p.d.f:Probability Density Function)とよばれる。
また、$F(x)=P(X\le x)$なる$F$を(累積)分布関数とよぶ。

### p.d.fの性質

+ $\displaystyle \int_{-\infty}^{\infty}f(x)\ge0$
+ $\displaystyle \int_{-\infty}^{\infty}f(x)dx=1$

### (累積)分布関数と密度関数の性質

+ $F(-\infty)=0$、$F(\infty)=1$
+ $F$は広義増加(非減少)関数。つまり常に、$x_1\le x_2$ならば$F(x_1)\le F(x_2)$となる。
+ $F$は右連続である
+ $\displaystyle F(\alpha<X\le\beta)=F(\alpha)-F(\beta)=\int_{\beta}^{\alpha}f(x)dx$

### 簡単な分布関数での定積分

#### 一様分布$U(a,b)$-この例の場合$U(-1,1)$

$$
f(x)=\left\{\begin{array}{ll}
\displaystyle \frac{1}{2} && (|x|\le 1)\\ 
0 && (|x|>1)
\end{array} 
\right.
$$

##### 実際
$$
\int_{-\infty}^{\infty}\frac{1}{2}dx=\int_{-1}^{1}\frac{1}{2}dx=\frac{1}{2}\left[x\right]_{-1}^{1}=\frac{1}{2}(1-(-1))=1
$$

##### $x<-1$
$f(x)=0$より
$$
F(x)=\int_{-\infty}^{x}f(y)dy=\int_{-\infty}^{x}0dy=0
$$

##### $-1<x<-1$
$f(x)=1/2$より
$$
F(x)=\int_{-\infty}^{x}f(y)dy=\int_{-1}^{x}\frac{1}{2}dy=\frac{1}{2}[y]_{-1}^{x}=\frac{1}{2}(x-(-1))=\frac{1}{2}(x+1)
$$

##### $1<x$
$-1<x<-1$の積分が寄与して、
$$
F(x)=\int_{-\infty}^{x}f(y)dy=\int_{-1}^{1}\frac{1}{2}dy=\frac{1}{2}\left[y\right]_{-1}^{1}=\frac{1}{2}(1-(-1))=1
$$

## $\Gamma$(ガンマ)関数-岩波数学入門コース 微分積分演習 和達三樹より


### 定義
$\Gamma$(ガンマ)関数は次の定積分で定義される

> $\displaystyle \Gamma(s)=\int_{0}^{\infty}x^{s-1}e^{-x}dx,\quad s>0$

性質の2つめを見てみると、$\Gamma$関数は階乗を整数以外に拡張したものと捉えられる。

### 性質-1回導出して、後は使いまわす!

+ $\displaystyle \Gamma(s+1)=s\Gamma(s)$。証明は教科書4章-積分法-章末問題-\[3\]\(1\)
+ とくに、$n$が負でない整数のとき、$\Gamma(n+1)=n!$。証明は教科書4章-積分法-章末問題-\[3\]\(1\)。
+ $\displaystyle \Gamma(1)=1$
+ $\displaystyle \Gamma(\frac{1}{2})=\sqrt{\pi}$
+ $\displaystyle \int_{0}^{\infty}x^{s-1}e^{-ax}dx=\frac{\Gamma(s)}{a^{s}}$

詳しくは[wikipedia:Gamma function](https://en.wikipedia.org/wiki/Gamma_function)(やはり日本語wikipediaは端折りすぎ)

## $B$(ベータ)関数-岩波数学入門コース 微分積分演習 和達三樹より


### 定義
$B$(ベータ)関数は次の定積分で定義される

> $\displaystyle B(p,q)=\int_{0}^{1}x^{p-1}(1-x)^{q-1}dx,\quad p,q>0$

### 性質-1回導出して、後は使いまわす!

+ $\displaystyle B(p,q)=B(q,p)$
+ $\displaystyle B(p,q)=\frac{\Gamma(p)\Gamma(q)}{\Gamma(p+q)}$
+ とくに、$p,q$が負でない整数のとき、$\displaystyle B(p,q)=\frac{(p-1)!(q-1)}{(p+q-1)!}$。なんとなく$\displaystyle {n \choose k}=\frac{n!}{k!(n-k)!}$の逆数っぽい?
+ $\displaystyle B(p,q)=\int_{0}^{\infty}\frac{x^{\alpha-1}}{(1+x)^{\alpha+\beta}}dx$

詳しくは[wikipedia:Beta function](https://en.wikipedia.org/wiki/Beta_function)(やはり日本語wikipediaは端折りすぎ)

## $\Gamma$関数と$B$関数の使いみち

### 期待値と分散

> 期待値 $\displaystyle \mu\overset{定義}{=}E[X]=\int_{-\infty}^{\infty}x\cdot f(x)dx$

> 分散  $\displaystyle \sigma^2\overset{定義}{=}V[x]=E[(X-\mu)^2]=\int_{-\infty}^{\infty}(x-\mu)^2f(x)dx$

### 指数分布$e(\sigma)$

#### 小寺平治 明解演習数理統計より
<img src="./img/lec05_exponentialDist.jpg" style="width: 600px">

#### もろもろの導出

##### $f(x)=e^{-ax},\quad (x>0)$として規格化定数は?

求めるには$f(x)=be^{-ax}$として実際に積分してみる。

$$
1=b\int_{0}^{\infty}e^{-ax}dx=b\int_{0}^{\infty}e^{-ax}dx
\overset{ax=tとおくと \\adx=dt}{=}\frac{b}{a}\int_{0}^{\infty}e^{-t}dt=\frac{b}{a}\int_{0}^{\infty}x^{1-1}e^{-t}dt=\frac{b}{a}\Gamma(1)=\frac{b}{a}
$$
より$b=a$となる。

分布関数は$\displaystyle f(x)=ae^{-ax},\quad (x>0)$の形をしている。

##### $a$の正体は?

ひとまず$\displaystyle f(x)=ae^{-ax},\quad (x>0)$の期待値を求めてみる。

$$
\mu=\int_{0}^{\infty}x\cdot ae^{-ax}dx\overset{ax=tとおくと \\adx=dt}{=}
\frac{1}{a}\int_{0}^{\infty}t^{2-1}e^{-t}dt=\frac{1}{a}\Gamma(2)=\frac{1}{a}\Gamma(1+1)=\frac{1}{a}1\cdot\Gamma(1)=\frac{1}{a}
$$

分散を求めてみる。

$$
\begin{eqnarray}
\sigma^2&=&E[(X-\mu)^2]=\int_{0}^{\infty}(x-\mu)^2e^{-ax}dx
=\int_{0}^{\infty}(x^2-2\mu x+\mu^2)e^{-ax}dx
\overset{ax=tとおくと \\adx=dt}{=}\int_{0}^{\infty}\left\{\frac{t^2}{a^2}-2\frac{\mu}{a}t +\mu^2\right\}e^{-t}dx \\
&=&\frac{1}{a^2}\Gamma(3)-2\frac{\mu}{a}\Gamma(2)+\mu^2\Gamma(1)=\frac{2}{a^2}-2\frac{\mu}{a}+\mu^2=\frac{2}{a^2}-\frac{2}{a^2}+\mu^2=\mu^2
\end{eqnarray}
$$

したがって、$\sigma=\mu$。

以上から、改めて$\displaystyle a=\frac{1}{\sigma}$として

$$
f(x)=\frac{1}{\sigma}e^{-\frac{x}{\sigma}}
$$

### 都合のいい密度関数$xe^{-x}$

$$
f(x)=\left\{\begin{array}{ll}
\displaystyle xe^{-x} && (x\ge0)\\ 
0 && (x<0)
\end{array} 
\right.
$$

<img src="./img/lec05_xe.jpg" style="width: 600px">

#### 規格化

$$
\int_{-\infty}^{\infty}f(x)dx=\int_{0}^{\infty}xe^{-x}dx=\int_{0}^{\infty}x^{2-1}e^{-x}dx=\Gamma(2)=\Gamma(1+1)=1\Gamma(1)=1
$$

#### 期待値

$$
\mu=\int_{-\infty}^{\infty}x\cdot f(x)dx=\int_{0}^{\infty}x\cdot xe^{-x}dx
=\int_{0}^{\infty}x^2e^{-x}dx=\Gamma(3)=\Gamma(2+1)=2\Gamma(1)=2
$$

### 都合のいい密度関数$6x(1-x),\quad(0\le x\le 1)$

<img src="./img/lec05_x_1-x_00.jpg" style="width: 600px">

#### 規格化

$[0,1]$区間で実際積分してみると
$$
\int_{0}^{1}6x(1-x)dx=6\int_{0}^{1}x^{2-1}(1-x)^{2-1}dx=6B(2,2)=6\frac{\Gamma(2)\Gamma(2)}{\Gamma(4)}=6\frac{1!1!}{3!}=1
$$

#### 期待値

$$
\mu=\int_{0}^{1}x\cdot6x(1-x)dx=6\int_{0}^{1}x^{3-1}(1-x)^{2-1}dx=6B(3,2)=6\frac{\Gamma(3)\Gamma(2)}{\Gamma(5)}=6\frac{2!1!}{4!}=\frac{1}{2}
$$
